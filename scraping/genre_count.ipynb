# Import Library

```python
import pandas as pd
import json
```

# Opening existing dataset

```python
df = pd.read_csv('../dataset/anime.csv')
```

# Exploring genres

```python
df['genres'].value_counts()
```

```python
genres = df['genres'].dropna()
```

```python
df_exploded = df.assign(genres=genres.str.split(';')).explode('genres')
df_exploded
```

```python
df_pivot = pd.crosstab(df_exploded['year'], df_exploded['genres'])
df_pivot = df_pivot.reset_index()
df_pivot
```

# Export genre time series

```python
df_pivot.to_csv('../dataset/genre_time_series.csv', index=False)
```

# Export genre list

```python
genre_list = list(set([splitted 
              for genre in set(genres) 
              for splitted in (genre.split(';') if ';' in genre else [genre])]))
genre_list
```

```python
out_file = open("../dataset/genres.json", "w")
json.dump(genre_list, out_file, indent = 4)
out_file.close()
```
